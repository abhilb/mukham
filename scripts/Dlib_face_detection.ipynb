```python
import imageio
import cv2
import dlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from pathlib import Path
```

```python
cur_dir = Path.cwd()
video_dir = cur_dir.parent / 'build'
video_fname = video_dir / 'demo2.mp4'
video_fname.exists()
```

```python
vid = cv2.VideoCapture(str(video_fname))
```

```python
_, frame = vid.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame)
plt.show()
```

```python
detector = dlib.get_frontal_face_detector()
```

```python
rects = detector(frame)
rects
```

```python
plt.imshow(frame)
for r in rects:
    plt.gca().add_patch(Rectangle((r.left(), r.top()),r.width(),r.height(),
                    edgecolor='red',
                    facecolor='none',
                    lw=4))
```

```python
frame.shape
```

```python
square_img = cv2.resize(frame, (512, 512), cv2.INTER_LINEAR)
plt.imshow(square_img)
plt.show()
```

```python
rects = detector(square_img)
rects
```

```python
small_img = cv2.resize(frame, None, None, 0.5, 0.5, cv2.INTER_LINEAR)
plt.imshow(small_img)
plt.show()
```

```python
rects = detector(small_img)
rects
```

```python
plt.imshow(small_img)
for r in rects:
    plt.gca().add_patch(Rectangle((r.left(), r.top()),r.width(),r.height(),
                    edgecolor='red',
                    facecolor='none',
                    lw=4))
```
